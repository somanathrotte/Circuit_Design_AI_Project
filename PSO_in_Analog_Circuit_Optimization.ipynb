```
import numpy as np
import matplotlib.pyplot as plt

# Define your objective function to be minimized
def objective_function(x):
    return x**2  # Replace this with your actual objective function

# PSO Parameters
num_particles = 20
num_dimensions = 1
max_iterations = 100
c1 = 2.0
c2 = 2.0
w = 0.5

# Initialization
particles_position = np.random.uniform(low=-5.0, high=5.0, size=(num_particles, num_dimensions))
particles_velocity = np.zeros_like(particles_position)
pbest_position = particles_position.copy()
pbest_value = np.full(num_particles, np.inf)
gbest_position = np.zeros(num_dimensions)
gbest_value = np.inf

# Lists to store the best fitness values over iterations
fitness_history = []

# PSO Main Loop with Plotting
for iteration in range(max_iterations):
    for i in range(num_particles):
        # Evaluate fitness
        fitness = objective_function(particles_position[i])

        # Update personal best
        if fitness < pbest_value[i]:
            pbest_value[i] = fitness
            pbest_position[i] = particles_position[i].copy()

        # Update global best
        if fitness < gbest_value:
            gbest_value = fitness
            gbest_position = particles_position[i].copy()

    for i in range(num_particles):
        # Update velocity and position
        r1, r2 = np.random.rand(), np.random.rand()
        cognitive_component = c1 * r1 * (pbest_position[i] - particles_position[i])
        social_component = c2 * r2 * (gbest_position - particles_position[i])
        particles_velocity[i] = w * particles_velocity[i] + cognitive_component + social_component
        particles_position[i] = particles_position[i] + particles_velocity[i]

    # Store the best fitness value in the current iteration
    fitness_history.append(gbest_value)

# Plot the convergence curve
plt.plot(range(max_iterations), fitness_history, label='Fitness Curve')
plt.title('PSO Convergence Curve')
plt.xlabel('Iteration')
plt.ylabel('Fitness Value')
plt.legend()
plt.show()

# Print the result
print("Optimal solution:", gbest_position)
print("Optimal value:", gbest_value)
```
